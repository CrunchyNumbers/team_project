```python
data_processed = pd.get_dummies(df, columns = ['Income_type' , 'Education_type', 'Family_status', 'Housing_type', 'Occupation_type'])
print(data_processed )
```

```python
y = data_processed['Target']
X = data_processed.drop(columns= ['Target'])
```

```python
X.info
```

```python
npx = np.array(X)
X_constant = sm.add_constant(npx)  # Add a constant term for the intercept
model = sm.OLS(y, X_constant.astype(float)).fit()

```

```python
print(model.summary)
```

```python
LRresult = (model.summary2().tables[1])
```

```python
print(LRresult.T)
```
